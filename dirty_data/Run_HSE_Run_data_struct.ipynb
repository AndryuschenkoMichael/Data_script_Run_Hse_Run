# Отформатровать таблицу rooms

```python
import pandas as pd
import numpy as np
```

Загрузка таблицы

```python
rooms = pd.read_csv("rooms.csv")
rooms.head(20)
```

Удаление ненужных столбцов

```python
rooms = rooms[['id', 'code', 'campus_id']]
rooms.head(20)
```

Выбрать из таблицы только те комнаты, которые находятся на Покровском бульваре(campus_id = 1)

```python
rooms = rooms[rooms['campus_id'] == 1]
rooms.head(20)
```

Сохраняем данные в rooms_for_vertices.csv для таблицы вершин

```python
rooms.to_csv('rooms_for_vertices.csv', index=False)
```

Выбрать только те аудитории, которые будут доступны в приложении "Run_HSE_Run"

```python
rooms_modified = rooms[rooms['code'].str.len() < 6]
```

Сохраняем данные в доступных в приложении аудиторий в rooms_modified.csv

```python
rooms_modified.to_csv('rooms_modified.csv', index=False)
```

# Отформатировать таблицу verticies

Загрузка таблицы

```python
vertices = pd.read_csv("vertices.csv")
vertices.head(20)
```

Удаление ненужных столбцов

```python
vertices = vertices[['id', 'type', 'room_id', 'campus_id']]
vertices.head(20)
```

Выбрать из таблицы только те вершины, которые находятся на Покровском бульваре(campus_id = 1)

```python
vertices = vertices[vertices['campus_id'] == 1]
vertices.head(20)
```

Заполним пропущенные значения значением: -1 в room_id

```python
vertices['room_id'].fillna(-1, inplace=True)
vertices['room_id'] = vertices['room_id'].astype(int)
vertices.head(20)
```

Сохраним даннные вершин в файл: vertices_for_edge.csv

```python
vertices.to_csv("vertices_for_edge.csv", index=False)
```

Выберем из таблицы только те данные, где type = roomDoor и в которых room_id соответствует какой либо комнате в таблице rooms_modified.csv

```python
vertices_modified = vertices[(vertices['type'] == "roomDoor") & (vertices['room_id'] != -1)]
array = np.array(rooms_modified['id'])
vertices_modified = vertices_modified.query("room_id in @array")
vertices_modified.head(20)
```

Сохраним данные в vertices_modified.csv

```python
vertices_modified.to_csv("vertices_modified.csv", index=False)
```
